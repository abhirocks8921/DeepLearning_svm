```python
x = df.drop(labels="1",axis=1)
y = df["1"]
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
s.fit(x)
x=s.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import Perceptron
model=Perceptron()
model.fit(x_train,y_train)
classifier_pred=model.predict(x_test)
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model, x, y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

recall=recall_score(y_test, classifier_pred, average='macro')
print("recall= ",recall)

accuracy=accuracy_score(y_test, classifier_pred)
print("accuracy= ",accuracy)

precision=precision_score(y_test, classifier_pred, average='macro')
print("precision= ",precision)
```

```python
grid = dict()
grid['eta0'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
search = GridSearchCV(model, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results = search.fit(x, y)
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```
